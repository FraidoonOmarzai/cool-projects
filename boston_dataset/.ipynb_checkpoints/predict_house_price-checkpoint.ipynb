```python
# we are gonna shuffle and split training and testing
prices = data['PRICE']
feature = data.drop('PRICE', axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(feature, prices,
                                                   test_size=0.2, random_state=10)

# % of training set
len(X_train)/len(feature)
```

```python
# % of test data set
X_test.shape[0]/feature.shape[0]
```

## Multivariable Regression

```python
reg = LinearRegression()
reg.fit(X_train, Y_train)

print('intercept: ', reg.intercept_)
print('training data r-squared: ', reg.score(X_train, Y_train))
print('testing data r-squared: ', reg.score(X_test, Y_test))


pd.DataFrame(data=reg.coef_, index=X_train.columns, columns=['coef'])
```

## Regression using log prices

```python
prices = np.log(data['PRICE'])
feature = data.drop('PRICE', axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(feature, prices,
                                                   test_size=0.2, random_state=10)

reg = LinearRegression()
reg.fit(X_train, Y_train)

print('intercept: ', reg.intercept_)
print('training data r-squared: ', reg.score(X_train, Y_train))
print('testing data r-squared: ', reg.score(X_test, Y_test))

pd.DataFrame(data=reg.coef_, index=X_train.columns, columns=['coef'])
```

```python
(np.e**0.080331) * 1000
```

## p values & Evaluating Coefficints

```python
x_incl_const = sm.add_constant(X_train)

model = sm.OLS(Y_train, x_incl_const)
result = model.fit()

# result.params
# result.pvalues

pd.DataFrame({'coef':round(result.params, 5), 'pvalues':round(result.pvalues,3)})
```

## Testing for Multicollinearity


```python
variance_inflation_factor(exog=x_incl_const.values, exog_idx=1)
```

```python
# len(x_incl_const.columns)
# or
x_incl_const.shape[1]
```

```python
vif=[]
for i in range(x_incl_const.shape[1]):
    vif.append(variance_inflation_factor(exog=x_incl_const.values, exog_idx=i))
print(vif)
```

```python
vif=[variance_inflation_factor(exog=x_incl_const.values, exog_idx=i)
                    for i in range(x_incl_const.shape[1])]
    
pd.DataFrame({"coef_name":x_incl_const.columns, "vif":np.around(vif,2)})

# if the value of vif<10 than no chance of multicollinearity
```

## Model Simplification & the BIC

```python
# original model with log prices and all features
x_incl_const = sm.add_constant(X_train)

model = sm.OLS(Y_train, x_incl_const)
result = model.fit()

# result.params
# result.pvalues

org_coef = pd.DataFrame({'coef':result.params, 'pvalues':round(result.pvalues,3)})

print('BIC is: ', result.bic)
print('r-squared is: ', result.rsquared)
```

```python
# reduced model #1 excluding INDUS, AGE
x_incl_const = sm.add_constant(X_train)
x_incl_const = x_incl_const.drop(['INDUS', 'AGE'], axis=1)

model = sm.OLS(Y_train, x_incl_const)
result = model.fit()

# result.params
# result.pvalues

coef_minus_indus = pd.DataFrame({'coef':result.params, 'pvalues':round(result.pvalues,3)})

print('BIC is: ', result.bic)
print('r-squared is: ', result.rsquared)
```

## Residuals & Residual Plots

```python
# modified model: transformed (using log prices) & simplified (dropping two features) 
prices = np.log(data['PRICE'])
feature = data.drop(['PRICE', 'INDUS', 'AGE'], axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(feature, prices,
                                                   test_size=0.2, random_state=10)

# using statsmodel
x_incl_const = sm.add_constant(X_train)
model = sm.OLS(Y_train, x_incl_const)
result = model.fit()

# residuals
# residuals = Y_train - result.fittedvalues
# residuals
# result.resid

# graph of acutual vs predicted prices
corr = round(Y_train.corr(result.fittedvalues), 2)

plt.figure(figsize=(12,10))
plt.scatter(Y_train, result.fittedvalues, c='navy', alpha=0.6)
plt.plot(Y_train, Y_train, c='cyan')
plt.title(f'Actual Vs Predicted log prices:(corr{corr})')
plt.xlabel('Acutal log prices $y_i$', fontsize=18)
plt.ylabel('Predicted log prices $\hat y_i$', fontsize=18)


plt.figure(figsize=(12,10))
plt.scatter(np.e**Y_train, np.e**result.fittedvalues, c='blue', alpha=0.6)
plt.plot(np.e**Y_train, np.e**Y_train, c='cyan')
plt.title(f'Actual Vs Predicted prices:(corr{corr})')
plt.xlabel('Acutal prices 000s $y_i$', fontsize=18)
plt.ylabel('Predicted prices 000s $\hat y_i$', fontsize=18)


# residuals vs predicted values
plt.figure(figsize=(12,10))
plt.scatter(result.fittedvalues, result.resid, c='skyblue', alpha=0.6)
plt.title('Residual Vs fitted values)')
plt.xlabel('Predicted log prices $\hat y_i$', fontsize=18)
plt.ylabel('Residual', fontsize=18)

plt.show()

# mean squared error and r-squared
reduced_log_mse = round(result.mse_resid, 3)
reduced_log_rsquared = round(result.rsquared, 3)
```

```python
# Distribution of residuals (log price)  - checking for normality
resid_mean = round(result.resid.mean(), 3)
resid_skew = round(result.resid.skew(), 3)

sns.distplot(result.resid, color='navy')
plt.title(f'log price model: residuals mean ({resid_mean}) skew ({resid_skew})')
plt.show()
```

```python
# modified model: transformed (using log prices) & simplified (dropping two features) 
prices = data['PRICE']
feature = data.drop(['PRICE'], axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(feature, prices,
                                                   test_size=0.2, random_state=10)

# using statsmodel
x_incl_const = sm.add_constant(X_train)
model = sm.OLS(Y_train, x_incl_const)
result = model.fit()

# graph of acutual vs predicted prices
corr = round(Y_train.corr(result.fittedvalues), 2)

plt.figure(figsize=(12,10))
plt.scatter(Y_train, result.fittedvalues, c='indigo', alpha=0.6)
plt.plot(Y_train, Y_train, c='cyan')
plt.title(f'Actual Vs Predicted prices:(corr{corr})')
plt.xlabel('Acutal prices $y_i$', fontsize=18)
plt.ylabel('Predicted prices $\hat y_i$', fontsize=18)

plt.show()

# residuals vs predicted values
plt.figure(figsize=(12,10))
plt.scatter(result.fittedvalues, result.resid, c='indigo', alpha=0.6)

plt.title('Residual Vs fitted values)')
plt.xlabel('Predicted log prices $\hat y_i$', fontsize=18)
plt.ylabel('Residual', fontsize=18)

plt.show()

# residuals Distribution chart
resid_mean = round(result.resid.mean(), 3)
resid_skew = round(result.resid.skew(), 3)

sns.distplot(result.resid, color='navy')
plt.title(f'residuals skew ({resid_mean}) mean ({resid_skew})')
plt.show()

# mean squared error and r-squared
full_normal_mse = round(result.mse_resid, 3)
full_normal_rsquared = round(result.rsquared, 3)
```

```python
# modified model: transformed (using log prices) & simplified (dropping two features) 
prices = np.log(data['PRICE'])
feature = data.drop(['PRICE', 'INDUS', 'AGE', 'LSTAT', 'RM', 'NOX', 'CRIM'], axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(feature, prices,
                                                   test_size=0.2, random_state=10)

# using statsmodel
x_incl_const = sm.add_constant(X_train)
model = sm.OLS(Y_train, x_incl_const)
result = model.fit()

# graph of acutual vs predicted prices
corr = round(Y_train.corr(result.fittedvalues), 2)

plt.figure(figsize=(12,10))
plt.scatter(Y_train, result.fittedvalues, c='#e74c3c', alpha=0.6)
plt.plot(Y_train, Y_train, c='cyan')
plt.title(f'Actual Vs Predicted prices:(corr{corr})')
plt.xlabel('Acutal log prices $y_i$', fontsize=18)
plt.ylabel('Predicted log prices with omitted variables', fontsize=18)

plt.show()

# residuals vs predicted values
plt.figure(figsize=(12,10))
plt.scatter(result.fittedvalues, result.resid, c='#e74c3c', alpha=0.6)

plt.title('Residual Vs fitted values')
plt.xlabel('Predicted log prices $\hat y_i$', fontsize=18)
plt.ylabel('Residual', fontsize=18)

plt.show()
 
    
# mean squared error and r-squared
omitted_var_mse = round(result.mse_resid, 3)
omitted_var_rsquared = round(result.rsquared, 3)
```

```python
pd.DataFrame({'MSE': [reduced_log_mse, full_normal_mse, omitted_var_mse],
             'R-Squared': [reduced_log_rsquared, full_normal_rsquared, omitted_var_rsquared],
              'RMSE':np.sqrt([reduced_log_mse, full_normal_mse, omitted_var_mse])},
            index = ['Reduced log model', 'Full normal price model', 'Omitted var model'])
```

```python
print('1 s.d. in log prices is ', np.sqrt(reduced_log_mse))
print('2 s.d. in log prices is ', 2*np.sqrt(reduced_log_mse))

upper_bound = np.log(30) + 2*np.sqrt(reduced_log_mse)
print('the upper bound in log prices for 95% prediction interval is ', upper_bound)
print('the upper bound in normal prices is $ ', np.e**upper_bound * 1000)

lower_bound = np.log(30) - 2*np.sqrt(reduced_log_mse)
print('the lower bound in log prices for 95% prediction interval is ', lower_bound)
print('the lower bound in normal prices is $ ', np.e**lower_bound * 1000)
```
